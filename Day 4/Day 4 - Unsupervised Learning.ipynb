```python
df = pd.read_csv("/kaggle/working/train.csv")
df
```

```python
df.isna().sum().describe()
df = df.dropna(axis=1)
df.describe()
```

```python
df_pca = df.drop(['date_time'],axis=1)
df_pca
```

```python
#check for optimal number of features
pca = PCA(random_state=123)
pca.fit(df_pca)
features = range(pca.n_components_)

plt.figure(figsize=(8,4))
plt.bar(features[:22], pca.explained_variance_[:22], color='lightskyblue')
plt.xlabel('PCA feature')
plt.ylabel('Variance')
plt.xticks(features[:15])
plt.show()
```

```python
df.plot(x = 'date_time', 
        y = 'price_usd', 
        figsize = (16, 8))

plt.xlabel('dates')
plt.ylabel('USD')
plt.title('Time series of room price by date of search');
```

```python
from sklearn.cluster import KMeans
ks = range(1, 10)
inertias = []

df = df[['price_usd','srch_booking_window','srch_saturday_night_bool']]

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df)
    inertias.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```
